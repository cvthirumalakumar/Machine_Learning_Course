## Question 4: Multi-layer Perceptrons

### Digit Classification

In this question, you will perform digit classification using MLP. You can use the MLPClassifier from sklearn. Train and two test sets have been provided [here](https://drive.google.com/drive/folders/1OUVrOMp2jSSBDJSqvEyXDFTrhiyZnqit?usp=sharing). Report the accuracy and any other interesting observations.

```python
import numpy as np
import h5py
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
```

```python
filename = "training_3digits.hdf5"
train = h5py.File(filename,'r')
train_images = np.array(train['images'])
train_digits = np.array(train['digits'])
train.close()
```

```python
filename = "testing_3digits_part1.hdf5"
test1 = h5py.File(filename,'r')
test_images_1 = np.array(test1['images'])
test_digits_1 = np.array(test1['digits'])
test1.close()
```

```python
filename = "testing_3digits_part2.hdf5"
test2 = h5py.File(filename,'r')
test_images_2 = np.array(test2['images'])
test_digits_2 = np.array(test2['digits'])
test2.close()
```

```python
mlp_classifier = MLPClassifier(hidden_layer_sizes=(100,), early_stopping=True, random_state=42)
mlp_classifier.fit(train_images.reshape(train_images.shape[0],-1), train_digits)
y_pred = mlp_classifier.predict(test_images_1.reshape(test_images_1.shape[0],-1))
accuracy = accuracy_score(test_digits_1, y_pred)
accuracy
```

```python
y_pred = mlp_classifier.predict(test_images_2.reshape(test_images_2.shape[0],-1))
accuracy = accuracy_score(test_digits_2, y_pred)
accuracy
```

```python
import matplotlib.pyplot as plt
rand_idx = np.random.choice(range(train_images.shape[0]), size=1)[0]
plt.imshow(train_images[rand_idx])
plt.title("random sample from train set with label "+str(train_digits[rand_idx]))
plt.show()
```

```python
import matplotlib.pyplot as plt
rand_idx = np.random.choice(range(test_digits_1.shape[0]), size=1)[0]
plt.imshow(test_images_1[rand_idx])
plt.title("random sample from test set 1 with label "+str(test_digits_1[rand_idx]))
plt.show()
```

```python
import matplotlib.pyplot as plt
rand_idx = np.random.choice(range(test_digits_2.shape[0]), size=1)[0]
plt.imshow(test_images_2[rand_idx])
plt.title("random sample from test set 2 with label "+str(test_digits_2[rand_idx]))
plt.show()
```

### Observations
