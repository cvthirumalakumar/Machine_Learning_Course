This question requires you to create a basic facial recognition system using a technique called principal component analysis (PCA)

The goal of implementing this system is to recognize a person's face by comparing it to a pre-existing database of faces, and identifying the closest match.

#### Implement PCA Algorithm.

Explain your steps with comments and write a brief explanation of the method.

```python
def principalComponentAnalysis(X):
    """
    Implement pca in the following function by taking the following arguements:
    1. X(images array)
    2. number of components
    The function must return the eigen vectors (principal components) also termed "Eigen faces"

    Explanation: The steps performed are calculating the eigen values and eigen vectors and taking eigen vectors corresponding to top_n eigen values
    as principal components. This method returns all eigne values and eigen vectors in a sorted order and the experiments includes changing n_components
    which is nothing but selecting top_n, procedure will be same. 
    """

    # your code here

    mean = np.mean(X, axis=0) # Computing mean face
    X = X - mean # Substracting mean face from the data to make to centered
    cov_matrix = (X.T @ X)/(X.shape[0]-1) # Calculating covariance matrix of the features
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix) # Calculating eigne values and eigen vectors of the covariance matrix
    eigenvectors = eigenvectors.T 
    idxs = np.argsort(eigenvalues)[::-1]  
    eigenvalues = eigenvalues[idxs]   #Sorting eigne values
    eigenvectors = eigenvectors[idxs]  #Soritng eigen vectors

    eigen_components = eigenvectors

    return np.real(eigen_components)
```

```python
# plot the eigen faces
%matplotlib inline
trainFaces_flattened = trainFaces.reshape(trainFaces.shape[0], -1)
eigenfaces = principalComponentAnalysis(trainFaces_flattened)
eigenfaces.shape
```

#### Implement Image Reconstruction from Eigenfaces


Explain your steps with comments and write a brief explanation of the method.

```python
def imageReconstruction(testFace, eigenFaces, meanFace):
    """
    Helper function to reconstruct images
    
    Explanation: the process involves first subtracting the mean face from the given face and then computing weights of 5 eigen faces
    as a dot product with each fo eigen face and adding 5 eigen faces with corresponding weights as a linear combination of eigen faces.
    And finally adding the mean face.
    """
    # your code here
    testFace_centered = testFace - meanFace   #subtracting mean face
    coefficients = np.dot(testFace.flatten(), eigenFaces.T)   #calculating weights associated with each eigen face
    reconstruction = np.dot(coefficients, eigenFaces)+ meanFace.flatten()   # reconstructing face as linear combination of eigen faces and adding mean face
    return reconstruction
```

```python
# plot recontructions here
%matplotlib inline
mean_face = np.mean(trainFaces, axis=0)
random_faces = []
count = 0
for i in np.unique(trainLabels):
    sub_faces = trainFaces[trainLabels==i]
    ranodm_idx = np.random.choice(range(sub_faces.shape[0]), size=1)
    random_face = sub_faces[ranodm_idx]
    random_faces.append(random_face)
    # print("Subject",i)
    reconstructed_face = imageReconstruction(random_face,eigenfaces[:100],mean_face)
    reconstructed_face = reconstructed_face.reshape(112, 92)
    fig, axes = plt.subplots(1, 2,figsize=(4, 2))
    axes[0].imshow(random_face.reshape(112,92),cmap='gray')
    axes[0].set_title('Original')
    axes[0].axis('off')
    axes[1].imshow(reconstructed_face,cmap='gray')
    axes[1].set_title('Reconstructed')
    axes[1].axis('off')
    if count == 3:
        break
    count+=1
    # break
```

#### Visualisation
Visualise the results for different number of factors(pc = 5, 10, 50, 100, etc.)

```python
%matplotlib inline

def displayNfactors(input_faces, pc=[5, 10, 50, 100, 500]):

    # your code here
    n = len(pc)
    count = 0
    for face in input_faces:
        reconstructed_faces_t = []
        for pc_t in pc:
            reconstructed_face = imageReconstruction(face,eigenfaces[:pc_t],mean_face)
            reconstructed_face = reconstructed_face.reshape(112, 92)
            reconstructed_faces_t.append(reconstructed_face)
        fig, axes = plt.subplots(1, n+1,figsize=(20, 4))
        axes[0].imshow(face.reshape(112,92),cmap='gray')
        axes[0].set_title('Original')
        axes[0].axis('off')
        for i in range(1,n+1):
            axes[i].imshow(reconstructed_faces_t[i-1].reshape(112,92),cmap='gray')
            axes[i].set_title(f'Reconstructed {pc[i-1]}')
            axes[i].axis('off')
        if count == 3:
            break
        count+=1
            
displayNfactors(random_faces)
```

#### Implement face recognition module based on the norm
Explain your steps with comments and write a brief explanation of the method.

*   Test the module and report accuracies based on the number of components taken for a range of value and plot them. 
*   Also plot the mean square error vs the number of eigenvectors taken and report your observations. 
*   For further empirical analysis, plot the semi-log variant of the error plot obtained above.

```python
def getClass(test_image, eigenfaces, meanface, train_faces, train_labels):
    """
    Arguments:
    1. Test image
    2. Eigenvectors
    3. Mean Eigen Face
    4. Training Set(X_train,y_train)

    The function must return the following:
    1. image with smallest error
    2. error value
    3. prediction class
    """

    # your code here
    reconstructed_test_image = imageReconstruction(test_image, eigenfaces, meanface)
    
    # Compute the error for each training image
    errors = []
    for i, train_image in enumerate(train_faces):
        reconstructed_train_image = imageReconstruction(train_image, eigenfaces, meanface)
        error = np.linalg.norm(reconstructed_train_image - reconstructed_test_image)
        errors.append(error)
    
    # Find the closest image and its error
    min_error= min(errors)
    prediction_class = train_labels[np.argmin(errors)]
    closest_image = train_faces[np.argmin(errors)]
    
    return closest_image, min_error, prediction_class
```

Iterate through all the images in the test data and test the accuracy by taking different number of components

```python
components = [3,5,10,20,30,40,50]
mse = []
accuracies = []
meanface = np.mean(trainFaces,axis = 0)
for numComponents in components:
    # your code here
    eigefaces_n = eigenfaces[:numComponents]
    errors = []
    preds = []
    for face in testFaces:
        closest_face, min_error, prediction_class = getClass(face, eigefaces_n, meanface, trainFaces, trainLabels)
        errors.append(min_error)
        preds.append(prediction_class)
    accuracies.append(accuracy_score(testLabels,preds))
    mse.append(sum(np.square(errors))/len(errors))
        
```

```python
plt.plot(components, accuracies)
plt.title("Number of eigenvectors vs accuracy")
plt.xlabel("Number of eigenvectors")
plt.ylabel("Accuracy")
plt.show()
```

Plot Number of eigenvectors vs Mean Square Error

```python
# your code here
plt.plot(components, mse)
plt.title("Number of eigenvectors vs Mean Square Error")
plt.xlabel("Number of eigenvectors")
plt.ylabel("MSE")
plt.show()
```

```python
# your code here
plt.plot(components, np.log(mse))
plt.title("Number of eigenvectors vs Mean Square Error")
plt.xlabel("Number of eigenvectors")
plt.ylabel("log-MSE")
plt.show()
```
