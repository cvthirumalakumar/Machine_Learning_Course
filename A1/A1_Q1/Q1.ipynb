```python
def label_encode(y):
    class_mapping = dict(zip(y.unique(),range(len(y.unique()))))
    return class_mapping, np.array([class_mapping[class_name] for class_name in y])

def shuffle_data(x,y):
    index_array = np.arange(len(x))
    np.random.shuffle(index_array)
    x_shuffled = x[index_array]
    y_shuffled = y[index_array]
    return x_shuffled, y_shuffled

def sample_dataset(x, y, train_split, test_split):
    x_train, y_train = np.empty((0,x.shape[-1])), np.empty(0,dtype=int)
    x_test, y_test = np.empty((0,x.shape[-1])), np.empty(0,dtype=int)
    for class_i in np.unique(y):
        class_i_idx = y == class_i
        x_class_i, y_class_i = x[class_i_idx], y[class_i_idx]
        x_class_i_shuffled, y_class_i_shuffled = shuffle_data(x_class_i, y_class_i)
        n_train, n_test = int(np.floor(len(y_class_i)*train_split)), int(np.floor(len(y_class_i)*test_split))
        x_train = np.concatenate((x_train, x_class_i_shuffled[:n_train]), axis=0)
        y_train = np.concatenate((y_train, y_class_i_shuffled[:n_train]), axis=0)

        x_test = np.concatenate((x_test, x_class_i_shuffled[n_train:n_train+n_test]), axis = 0)
        y_test = np.concatenate((y_test, y_class_i_shuffled[n_train:n_train+n_test]), axis = 0)
    return x_train, y_train, x_test, y_test
  
y_labels = data['Genus']
c_map, y = label_encode(y_labels)
x = data.iloc[:,:21].to_numpy()
```

# Part 1 : Decision Tree

```python
x_train, y_train, x_test, y_test = sample_dataset(x[:,:6], y, 0.2, 0.8)
```

```python
class DecisionTree:
    def __init__(self, max_depth=None):
        self.max_depth = max_depth
        self.tree_ = None
    def fit(self, X, y, depth=0):
        if len(set(y)) == 1 or depth == self.max_depth:
            return {'class': np.bincount(y).argmax()}

        best_feature, threshold = self.find_best_split(X, y)
        left_indices = X[:, best_feature] <= threshold
        right_indices = ~left_indices

        if not np.any(left_indices) or not np.any(right_indices):
            return {'class': np.bincount(y).argmax()}

        node = {'feature': best_feature, 'threshold': threshold}
        node['left'] = self.fit(X[left_indices], y[left_indices], depth + 1)
        node['right'] = self.fit(X[right_indices], y[right_indices], depth + 1)

        self.tree_ = node
        return node

    def find_best_split(self, X, y):
        num_features = X.shape[1]
        best_feature, best_threshold, best_score = None, None, 100
        for feature in range(num_features):
            thresholds = np.unique(X[:, feature])

            for threshold in thresholds:
                left_indices = X[:, feature] <= threshold
                right_indices = ~left_indices

                entropy_score = self.node_entropy(y, y[left_indices], y[right_indices])
                if entropy_score < best_score:
                    best_score = entropy_score
                    best_feature = feature
                    best_threshold = threshold
        return best_feature, best_threshold

    def node_entropy(self, parent, left_child, right_child):
        p = len(parent)
        pl = len(left_child)
        pr = len(right_child)

        entropy_score = (pl / p) * self.calculate_entropy(left_child) + (pr / p) * self.calculate_entropy(right_child)
        return entropy_score

    def calculate_entropy(self, labels):
        _, counts = np.unique(labels, return_counts=True)
        probabilities = counts / len(labels)
        entropy = -np.sum(probabilities * np.log2(probabilities))
        return entropy

    def predict_instance(self, x, node):
        if 'class' in node:
            return node['class']

        if x[node['feature']] <= node['threshold']:
            return self.predict_instance(x, node['left'])
        else:
            return self.predict_instance(x, node['right'])

    def predict(self, X):
        return [self.predict_instance(x, self.tree_) for x in X]
```

```python
tree_test_accuracies = []
for depth in range(1,16):
    print("Training Decision tree with depth :",depth, end='\r')
    tree = DecisionTree(max_depth=depth)
    tree.fit(x_train, y_train)
    predictions = tree.predict(x_test)
    tree_test_accuracies.append((sum(predictions == y_test)/len(y_test))*100)
```

```python
plt.plot(range(1,16),tree_test_accuracies)
plt.title('Decision Tree accuracy plot on test set')
plt.xlabel('Depth')
plt.ylabel('Accuracy (%)')
```

# Decision Tree Observations
1. As depth increasing, test set accuracy increases till depth = 8.
2. After depth = 8, a slight decrease in the accuracy till depth = 10 and then constant.
3. So, depth = 8 can be considered as the best choice for this dataset.
4. As dataset is not class balanced, test set also has the same class proportions. Hence accuracy may not be the oprimal performance measure and average F1 score can be well suited. Since training data also is class imbalanced, the model will be overfitting to the majority class. (Hypothesis only)
5. Experiments can be performed with F1 score to validate above points, but not performed in the interest of time and the same can be applicable for kNN also.

# Part 2 : k Nearest Neighbours

```python
x_train, y_train, x_test, y_test = sample_dataset(x, y, 0.3, 0.1)
```

```python
class KNN:
    def __init__(self, k=3):
        self.k = k
        self.X_train = None
        self.y_train = None

    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def calculate_distance(self, x1, x2):
        return np.sqrt(np.sum(np.square(np.absolute(x1-x2))))

    def predict_instance(self, x):
        distances = [self.calculate_distance(x, x_temp) for x_temp in self.X_train]
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = [self.y_train[i] for i in k_indices]
        prediction = np.bincount(k_nearest_labels).argmax()
        return prediction
    
    def predict(self, X_test):
        predictions = [self.predict_instance(x) for x in X_test]
        return np.array(predictions)
```

```python
knn_test_accuracies = []
for k in range(1, 51):
    print("KNN with k = ", k, end='\r')
    knn = KNN(k=k)
    knn.fit(x_train, y_train)
    predictions = knn.predict(x_test)
    knn_test_accuracies.append((sum(predictions == y_test)/len(y_test))*100)
```

```python
plt.plot(range(1,51),knn_test_accuracies)
plt.title('kNN accuracy plot on test set')
plt.xlabel('k')
plt.ylabel('Accuracy (%)')
```

# kNN Observations
1. As k increases the performance (Test set accuracy) starts decreasing.
2. Rights from k = 1, downfall is observed and the best accuracy is obtained at k = 1.

```python
x_train, y_train, x_test, y_test = sample_dataset(x, y, 0.8, 0.2)
```

```python
plt.plot(range(1,16),tree_test_accuracies_sk)
plt.title('Scikit-Learn Decision Tree accuracy plot on test set')
plt.xlabel('Depth')
plt.ylabel('Accuracy (%)')
```

## 3.2 kNN
